```python
# 한글 폰트 설정
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic' # windows용 -> 맑은 고딕체 
matplotlib.rcParams['font.size']=15 # 글자 크기 설정 
```

```python
matplotlib.rcParams['font.family']='HYGungSo-Bold' # windows용 -> 굵은 궁서체
matplotlib.rcParams['font.size']=20 # 글자 크기 설정 
```

```python
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용 시에 마이너스 깨짐 오류 해결 
plt.plot([-1,0,1],[-5,-1,2])
plt.xlabel('X축',color='red')
plt.ylabel('Y축',color='#00aa00')
```

```python
# 축 제목 위치 변경하기 
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15 
plt.title('꺾은선 그래프')
matplotlib.rcParams['axes.unicode_minus']=False # 한글 폰트 사용 시에 마이너스 깨짐 오류 해결 
plt.plot([1,2,3],[2,4,8])
plt.xlabel('X축',color='red', loc='right') # loc: 위치 지정 (top, bottom, left, right, center)
plt.ylabel('Y축',color='#00aa00', loc='top')
```

```python
# 축 label 
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['font.size']=15 
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
# 축 색상 설정하기 
plt.plot(x,y)
plt.xlabel('X축',color='red')
plt.ylabel('Y축',color='#00aa00')
```

```python
# 축 제목 위치 변경
plt.plot(x,y)
plt.xlabel('X축',color='red', loc='right')
plt.ylabel('Y축',color='#00aa00', loc='top')
```

```python
# Tick 조정하기 
# 틱 (Tick)은 그래프의 축에 간격을 구분하기 위해 표시하는 눈금이다.
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
```

```python
# 범례: label, legend
plt.plot(x,y,label='테스트')
plt.legend()
```

```python
# 범례 위치 조정하기
plt.plot(x,y,label='테스트')
plt.legend(loc='upper right') # upper, lower, left, right 조합하여 위치 변경 가능함.
```

```python
# 범례 위치 조정하기 2 
plt.plot(x,y,label='범례')
plt.legend(loc=(0.5,0.5)) # 튜플 형태로 0~1 사이 세부 지정 가능
```
