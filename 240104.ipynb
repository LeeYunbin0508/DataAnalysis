```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False


area = ['서울','인천','경기','강원','대전','충북','충남','세종','광주','전북','전남','대구','경북','부산','울산','경남','제주']
amt = [72915,14931,93013,5264,5789,6726,9915,1117,7318,6109,5429,8945,10047,13760,5397,13872,4661]
cnt = [1193368,305127,1696712,130160,128455,148354,199980,22645,131440,141288,112608,196122,211877,301430,104568,304286,74292]
plt.plot(amt)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(amt,'bo-.')
plt.title('지역별 체납액',fontsize=15)
plt.xlabel('지역')
plt.ylabel('체납금액')
plt.xticks(range(len(amt)), labels=area, rotation=45)
plt.grid(True, color='gray', alpha=0.5, ls='--')
plt.axhline(y=20000, color='red', ls='--', lw=0.5)
for i in range(len(amt)):
    plt.text(i-0.2, amt[i]+3000,amt[i]) # X, Y, 값 
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,1,1) # 2행 1열 1번째
plt.plot(area, amt, 'go--')
plt.title('체납액')
plt.subplot(2,1,2) # 2행 1열 2번째
plt.plot(area,cnt,'bx-')
plt.title('체납건수')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1) # 1행 2열 1번째
plt.plot(area, amt, 'go--')
plt.title('체납액')
plt.subplot(1,2,2) # 1행 2열 2번째
plt.plot(area,cnt,'bx-')
plt.title('체납건수')
plt.show()
```

```python
plt.figure(figsize=(10,8)) # 가로 10센치 세로 8센치인 그림을 생성
plt.plot(area, amt, 'go--') # x축에 area, y축에 amt 데이터를 초록색 동그라미와 점선으로 표시
plt.title('체납액') # 그래프에 '체납액'이라는 제목을 부여
plt.xticks(range(len(amt)), labels=area, rotation=45) # x축의 눈금을 설정
plt.show()
```

```python
# 체납 현황 분석 
plt.figure(figsize=(10,8))
plt.plot(area, amt, 'go--', label='체납액') # x축을 area, y축을 amt로하여 범례 제목을 '체납액', 녹색, 원형의 -- 모양 직선 그래프를 그림
plt.plot(area, cnt, 'bx-', label='체납건수') # x축을 area, y축을 cnt로 하여 범례 제목을 '체납건수', 파란색, x형 - 모양 직선 그래프를 그림
plt.title('체납액') # 그래프 제목을 '체납액'으로 함
plt.xticks(range(len(amt)), labels=area, rotation=45) # x축 눈금을 설정함
plt.legend() # 범례 표시 
plt.show()
```

```python
# 다양한 그래프 그리기 

# 막대 그래프 그리기 
plt.figure(figsize=(12,8)) # 가로 12인치, 세로 8인치의 그래프를 그림
plt.subplot(2,3,1) # 2행 3열의 1번째 그래프
plt.bar(area,amt) # x축을 area, y축을 amt로하여 막대 그래프를 그림
plt.title('막대 그래프') # 그래프 제목을 '막대 그래프'로 함 

# 가로막대 그래프 그리기
plt.subplot(2,3,2) # 2행 3열의 2번째 그래프
plt.barh(area,amt) # x축을 area, y축을 amt로하여 가로막대 그래프를 그림
plt.title('가로막대 그래프') # 그래프 제목을 '가로막대 그래프'로 함

# 산점도 그래프 그리기
plt.subplot(2,3,3) # 2행 3열의 3번째 그래프
plt.scatter(cnt,amt) # x축을 cnt, y축을 amt로하여 산점도 그래프를 그림
plt.title('산점도 그래프')

# 파이 그래프 그리기
plt.subplot(2,3,4) # 2행 3열의 4번째 그래프
plt.pie(amt) # amt에 대한 파이 그래프를 그림 
plt.title('파이 그래프')

# 히스토그램 그리기
plt.subplot(2,3,5) # 2행 3열의 5번째 그래프
plt.hist(amt) # amt에 대한 히스토그램을 그림
plt.title('히스토그램')

# 박스플롯 그리기
plt.subplot(2,3,6) # 2행 3열 6번째 그래프
plt.boxplot(amt) # amt에 대한 박스플롯을 그림
plt.title('박스플롯')

plt.show()
```
